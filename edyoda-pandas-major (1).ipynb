```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv")
```

```python
# Find no. of rows & columns in the dataset
print("rows :",df.shape[0])
print("columns :",df.shape[1])
```

```python
# Data types of columns.
df.dtypes
```

```python
# Info & describe of data in dataframe.
df.info()
```

```python
df.describe().round(2)
```

```python
# Find count of unique values in location column.
len(df.location.unique())
```

```python
# Find which continent has maximum frequency using values
max_freq = df.continent.value_counts().max()
con = df.continent.value_counts().index[0]

print("Continent with maximum frequency :")
print(con,":",max_freq)
```

```python
# Find maximum & mean value in 'total_cases'.
print("maximum :",df.total_cases.max())
print("Mean value :",df.total_cases.mean())
```

```python
# Find 25%,50% & 75% quartile value in 'total_deaths'.
print("25% quartile value :",df.total_deaths.quantile(.25))
print("50% quartile value :",df.total_deaths.quantile(.5))
print("75% quartile value :",df.total_deaths.quantile(.75))
```

```python
# Find which continent has maximum 'human_development_index'.
continent = df.groupby("continent").mean()

continent.loc[continent.human_development_index == continent.human_development_index.max(),['human_development_index']]
```

```python
# Find which continent has minimum 'gdp_per_capita'.
continent.loc[continent.gdp_per_capita == continent.gdp_per_capita.min(),['gdp_per_capita']]
```

```python
# Filter the dataframe with only this columns['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] and update the data frame.
df = df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
df
```

```python
# Remove all duplicates observations
duplicate_df = df[df.duplicated()]
duplicate_df
# no duplicate values
```

```python
# Find missing values in all columns
df.isnull().sum()
```

```python
# Remove all observations where continent column value is missing
df.dropna(subset = ['continent'], inplace = True)
df.isnull().sum()
```

```python
# Fill all missing values with 0
df.fillna(0, inplace  = True)
```

```python
# Convert date column in datetime format using pandas.to_datetime
df['date'] = pd.to_datetime(df.date)
```

```python
# Create new column month after extracting month data from date column.
df['month'] = df['date'].dt.month
```

```python
# Find max value in all columns using groupby function on 'continent' column.
# Store the result in a new dataframe named 'df_groupby'.
df_groupby = df.groupby("continent").max()
df_groupby.reset_index(inplace = True)
```

```python
# Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
df_groupby['total_deaths_to_total_cases'] = df_groupby.total_deaths/df_groupby.total_cases
```

```python
# Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
plt.figure(figsize = (6,5))
sns.set_style("darkgrid")
sns.distplot(x = df['gdp_per_capita'], bins = 50)
```

```python
# Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
df[gdp_per_capita] = df
plt.figure(figsize = (10,8))
sns.scatterplot(x = df["gdp_per_capita"], y = df["total_cases"], hue = df["continent"])
```

```python
# Plot Pairplot on df_groupby dataset.
sns.pairplot(df_groupby,hue = "continent")
```

```python
# Plot a bar plot of 'continent' column with 'total_cases' .
sns.catplot(x = "continent", y = "total_cases", kind = "bar", data = df_groupby)
```
